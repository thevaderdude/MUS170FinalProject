```python
# Spectrum data is pretty messy, might have to resort to rnn method, but first let's try my peak detection algorithm
def get_levels_max(height, num):
    key_levels = []
    key_idx = []
    bounded_idx = [0]
    num_iter = 0
    height2 = np.copy(height)
    while len(key_levels) < num:
        # get max index of height array
        max_idx = np.argmax(height2)

        # check bounded idxs to see if it is next to a peak
        on_bounded = False
        for b in bounded_idx:
            if max_idx + 1 == b or max_idx - 1 == b or max_idx == b:
                on_bounded = True

        # if its not bounded add to key levels
        if not on_bounded:
            key_levels.append(height[max_idx])
            key_idx.append(max_idx)
        # either way the idx is added to the bounded if its a key level or not
        bounded_idx.append(max_idx)

        # remove from relevance
        height2[max_idx] = 0
        num_iter += 1
        if num_iter > height2.shape[0]:
            # print('brih')
            break
    return key_levels, key_idx

t_levels, t_idx = get_levels_max(np.abs(trumpet_fft_out), 40)
t_maxes = np.array([trumpet_freqs[i] for i in t_idx])
t_levels, t_maxes
```

```python
# data creation method:
def generate_input_data(filepath, samp_size=2000, num_harmonics=20, freq_scaler=None):
    # Divide audio into samples of size 2000
    sample_rate, samples = wavfile.read(audio_path + filepath)
    samples_list = []
    for i in range(0, len(samples), samp_size):
        samples_list.append(samples[i:i+samp_size])
    # Run data through fft
    fft_outs = []
    freqs = fftfreq(samp_size, 1.0/sample_rate)
    for s in samples_list:
        fft_outs.append(fft(s))
    x_levels = []
    x_freqs = []
    # Run peak detection algo on data (n = 40)
    for f in fft_outs:
        levels, idx = get_levels_max(np.abs(f), num_harmonics*2)
        maxes = np.array([freqs[i] for i in idx])
        # Add non-repeat frequencies and corrosponding levels into data (size: 20 f + 20 l = 40)
        levs, fs = [], []
        for l, m in zip(levels, maxes):
            if len(levs) == num_harmonics:
                break
            if abs(m) not in fs:
                levs.append(l)
                fs.append(abs(m))
        # min-max scale levels data within each sample
        levs = np.array(levs)
        scaler = MinMaxScaler()
        scaler.fit(levs.reshape(-1, 1))
        levs = scaler.transform(levs.reshape(-1, 1))
        # make sure that input is valid size
        if len(levs) == num_harmonics:
            x_levels.append(levs)
            x_freqs.append(fs)
    # min-max scale frequencies on entire dataset (analyze min and max first)
    x_freqs = np.array(x_freqs)
    x_levels = np.array(x_levels)
    f_scaler = 0
    if freq_scaler is not None:
        f_scaler = freq_scaler
    else: 
        f_scaler = MinMaxScaler()
        f_scaler.fit(x_freqs.flatten().reshape(-1, 1))
    for i in range(len(x_freqs)):
        x_freqs[i] = f_scaler.transform(x_freqs[i].reshape(-1, 1)).reshape(num_harmonics)
    # create final input array
    x = np.zeros((len(x_levels), num_harmonics*2))
    for i in range(len(x_levels)):
        for j in range(num_harmonics):
            x[i, j] = x_levels[i, j]
            x[i, j+num_harmonics] = x_freqs[i, j]
    return x, f_scaler
    
```

```python
# 0 = piano, 1 = trumpet
x_p, f_scaler_t = generate_input_data('PianoChromatic.wav')
y_p = np.zeros(x_p.shape[0])
```

```python
f_scaler_t.data_max_
```

```python
x_t, _ = generate_input_data('TrumpetChromatic.wav', freq_scaler=f_scaler_t)
y_t = np.ones(x_t.shape[0])
x_t.shape
```

```python
x = np.concatenate((x_t, x_p))
y = np.concatenate((y_t, y_p))
x.shape
```

```python
# shuffle arrrays
p = np.random.permutation(len(x))
x, y = x[p], y[p]
```

```python
# split into train and test
test_partition = len(x) - 200
x_train = x[:test_partition]
x_test = x[test_partition:]
y_train = y[:test_partition]
y_test = y[test_partition:]
```

```python
clf = MLPClassifier(alpha=1e-5, hidden_layer_sizes=(32, 24), random_state=420, solver='lbfgs', max_iter=50_000)
clf.fit(x_train, y_train)
```

```python
train_results = clf.predict(x_train)
accuracy_score(y_train, train_results)
```

```python
test_results = clf.predict(x_test)
accuracy_score(y_test, test_results)
```

```python
def predict_instrument_from_file(filepath, model=clf, freq_scaler=f_scaler_t):
    data, f_scaler = generate_input_data(filepath, freq_scaler=freq_scaler)
    pred = clf.predict(data)
    proba = np.mean(pred)
    inst = 'Trumpet'
    if proba < 0.5:
        inst = 'Piano'
        proba = 1 - proba
    return inst, proba
```
